```python
# Original, raw data
df_org = pd.read_csv('dataset_olympics.csv')
df_org.head()
```

```python
# Data Cleaning
```

```python
def data_cleaning(df_org, sex, season):
  df = df_org[df_org['Sex'] == sex]                                     # Filter by sex
  df = df[df['Season'] == season]                                       # Filter by season
  df = df.drop(['Sex', 'Season', 'Games'], axis = 1)                    # Unnecessary since already filtered; games is redundant of "Year" column
  df['Weight'].fillna(df['Weight'].median(), inplace=True)              # NaN Weight values
  df['Age'].fillna(df['Age'].median(), inplace=True)                    # NaN Age values
  df['Height'].fillna(df['Height'].median(), inplace=True)              # NaN Height values
  df = df.drop(['ID', 'City', 'Team'], axis = 1)                        # Redundant: ID and Name of the athletes, City with the Year, Team and NOC
  df['Medal'] = df['Medal'].apply(lambda x: 0 if pd.isnull(x) else 1)   # Outcome variable: 1 for winning any medal, 0 for not winning

  one_hot_encoded = pd.get_dummies(df['Name'], prefix='Name')           # Dummy for each athletes
  df = pd.concat([df, one_hot_encoded], axis=1).drop('Name', axis = 1)
  one_hot_encoded = pd.get_dummies(df['NOC'], prefix='NOC')             # Dummy for each NOC
  df = pd.concat([df, one_hot_encoded], axis=1).drop('NOC', axis = 1)
  one_hot_encoded = pd.get_dummies(df['Year'], prefix='Year')           # Dummy for each year the Olympic was held
  df = pd.concat([df, one_hot_encoded], axis=1).drop('Year', axis = 1)
  one_hot_encoded = pd.get_dummies(df['Sport'], prefix='Sport')         # Dummy for sports
  df = pd.concat([df, one_hot_encoded], axis=1).drop('Sport', axis = 1)
  one_hot_encoded = pd.get_dummies(df['Event'], prefix='Event')         # Dummy for events
  df = pd.concat([df, one_hot_encoded], axis=1).drop('Event', axis = 1)
  return df
```

```python
# Female Summer
df_fs = data_cleaning(df_org, 'F', 'Summer')
X_fs = df_fs.drop(['Medal'], axis = 1)
y_fs = df_fs['Medal']
df_fs.head()
```

```python
# Female Winter
df_fw = data_cleaning(df_org, 'F', 'Winter')
X_fw = df_fw.drop(['Medal'], axis = 1)
y_fw = df_fw['Medal']
df_fw.head()
```

```python
# Male Summer
df_ms = data_cleaning(df_org, 'M', 'Summer')
X_ms = df_ms.drop(['Medal'], axis = 1)
y_ms = df_ms['Medal']
df_ms.head()
```

```python
# Male Winter
df_mw = data_cleaning(df_org, 'M', 'Winter')
X_mw = df_mw.drop(['Medal'], axis = 1)
y_mw = df_mw['Medal']
df_mw.head()
```

```python
# Data cleaning completed
# Ensemble method: Random Forest, MLP Classifier, and Decision Tree => Take the majority vote
```

```python
def get_accuracy_score(X, y):

  clf = DecisionTreeClassifier()
  mlp_model = MLPClassifier(hidden_layer_sizes=(2, 2), max_iter=5, random_state=0)
  forest_model = RandomForestClassifier(n_estimators=10)

  X_train, X_val, Y_train, Y_val = train_test_split(X,y)

  clf.fit(X_train, Y_train)
  mlp_model.fit(X_train, Y_train)
  forest_model.fit(X_train, Y_train)

  clf_y_pred = clf.predict(X_val)
  mlp_y_pred = mlp_model.predict(X_val)
  forest_y_pred = forest_model.predict(X_val)

  stacked_arrays = np.column_stack((clf_y_pred, mlp_y_pred, forest_y_pred))
  majority_y_pred, counts = mode(stacked_arrays, axis=1)

  score = accuracy_score(majority_y_pred, Y_val)
  return score
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import StackingClassifier

def V2_get_accuracy_score(X, y):
    # create estimators
    estimators = [
            ('DTC',DecisionTreeClassifier())
            ,('mlp',MLPClassifier(hidden_layer_sizes=(2, 2), max_iter=10, random_state=0))
            ,('RFC',RandomForestClassifier(n_estimators=10))]
    # Store in StackingClassifier constructor - uses LogisticRegression for voting
    clf = StackingClassifier(
            estimators=estimators, final_estimator=LogisticRegression()
            )


    X_train, X_val, Y_train, Y_val = train_test_split(X,y)

    clf.fit(X_train, Y_train)
    voted_prediction = clf.predict(X_val)
    score = accuracy_score(voted_prediction, Y_val)

    return score
```
